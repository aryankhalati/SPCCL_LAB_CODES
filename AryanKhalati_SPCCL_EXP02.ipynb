```python

assembly_code = [
    "START 100",
    "LOOP: ADD A, B",
    "SUB C, D",
    "MOVER A, B",
    "LOOP2: MULT A, C",
    "STOP",
    "END"
]

LC = 0
symbol_table = {}
intermediate_code = []

print("LC\tLabel\tOpcode\tOperands")

for line in assembly_code:
    parts = line.replace(',', '').split()

    label = None
    opcode = None
    operands = []

    if ":" in parts[0]:
        label = parts[0].replace(":", "")
        opcode = parts[1] if len(parts) > 1 else None
        operands = parts[2:]

        symbol_table[label] = LC
    else:
        opcode = parts[0]
        operands = parts[1:]

    if opcode == "START":
        LC = int(operands[0])
        intermediate_code.append((LC, "AD", "START", operands))
        print(f"{LC}\t{label}\t{opcode}\t{operands}")
        continue

    if opcode == "END":
        intermediate_code.append((LC, "AD", "END", operands))
        print(f"{LC}\t{label}\t{opcode}\t{operands}")
        break

    if opcode == "STOP":
        intermediate_code.append((LC, "IS", opcode, operands))
    else:
        intermediate_code.append((LC, "IS", opcode, operands))

    print(f"{LC}\t{label}\t{opcode}\t{operands}")

    LC += 1


print("\nSymbol Table:")
print("Label\tAddress")
for sym, addr in symbol_table.items():
    print(f"{sym}\t{addr}")

print("\nIntermediate Code:")
print("LC\tClass\tOpcode\tOperands")
for code in intermediate_code:
    print(f"{code[0]}\t{code[1]}\t{code[2]}\t{code[3]}")
```
