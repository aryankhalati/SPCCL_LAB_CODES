```python
def generate_machine_code(three_address_code):

    machine_instructions = []
    register_count = 0

    print(f"{'3AC Instruction':<20} | {'Generated Assembly'}")
    print("-" * 50)

    for target, op1, operator, op2 in three_address_code:

        r1 = f"R{register_count}"
        r2 = f"R{register_count + 1}"
        r3 = f"R{register_count + 2}"


        asm = [
            f"MOV {r1}, {op1}",
            f"MOV {r2}, {op2}",
        ]

        if operator == '+':
            asm.append(f"ADD {r3}, {r1}, {r2}")
        elif operator == '*':
            asm.append(f"MUL {r3}, {r1}, {r2}")

        asm.append(f"STR {target}, {r3}")


        asm_string = " ; ".join(asm)
        print(f"{target + ' = ' + op1 + ' ' + operator + ' ' + op2:<20} | {asm_string}")

        machine_instructions.extend(asm)
        register_count += 3
    return machine_instructions


tac_input = [
    ("t1", "x", "+", "y"),
    ("t2", "t1", "*", "z")
]

print("Starting Code Generation Phase...\n")
assembly_output = generate_machine_code(tac_input)
```
