```python

assembly_code = [
    "MACRO",
    "INCR X",
    "ADD X, =1",
    "MEND",
    "START 100",
    "INCR A",
    "INCR B",
    "STOP",
    "END"
]

MDT = {}
output_code = []

i = 0
n = len(assembly_code)

while i < n:
    line = assembly_code[i].strip()
    parts = line.split()

    if line == "MACRO":
        i += 1
        header = assembly_code[i].split()
        macro_name = header[0]
        parameters = header[1:]

        macro_body = []
        i += 1

        while assembly_code[i].strip() != "MEND":
            macro_body.append(assembly_code[i])
            i += 1

        MDT[macro_name] = (parameters, macro_body)

    else:
        parts = line.replace(',', '').split()

        if parts[0] in MDT:
            macro_name = parts[0]
            actual_args = parts[1:]

            parameters, macro_body = MDT[macro_name]

            param_map = dict(zip(parameters, actual_args))

            for stmt in macro_body:
                expanded = stmt
                for p in param_map:
                    expanded = expanded.replace(p, param_map[p])
                output_code.append(expanded)

        else:
            output_code.append(line)

    i += 1

print("Macro Definition Table (MDT):")
for name, (params, body) in MDT.items():
    print(f"\nMacro Name: {name}")
    print(f"Parameters: {params}")
    print("Body:")
    for stmt in body:
        print(stmt)

print("\nExpanded Assembly Code:")
for line in output_code:
    print(line)
```
