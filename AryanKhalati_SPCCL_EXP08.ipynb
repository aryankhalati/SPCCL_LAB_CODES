```python
import json

class ActivationRecord:
    def __init__(self, func_name, params, locals_dict, return_to):
        self.func_name = func_name
        self.params = params
        self.locals_dict = locals_dict
        self.return_to = return_to

    def __repr__(self):

        return json.dumps({
            "Function": self.func_name,
            "Parameters": self.params,
            "Locals": self.locals_dict,
            "Return Link": self.return_to
        }, indent=4)

class CallStack:
    def __init__(self):
        self.stack = []

    def push(self, record):
        print(f"\n---> [PUSH] Calling {record.func_name}")
        self.stack.append(record)
        self.display_stack()

    def pop(self):
        if self.stack:
            removed = self.stack.pop()
            print(f"\n<--- [POP] Returning from {removed.func_name}")
            self.display_stack()
            return removed
        print("Stack Underflow")
        return None

    def display_stack(self):
        print("---------- CURRENT CALL STACK ----------")
        if not self.stack:
            print("Empty")
        else:
            for i, record in enumerate(reversed(self.stack)):
                print(f"Level {len(self.stack) - i}: {record.func_name}")
        print("----------------------------------------")



call_stack = CallStack()

def simulate_factorial(n, caller="Main"):

    local_vars = {"n": n}
    current_record = ActivationRecord(f"Factorial({n})", {"input": n}, local_vars, caller)
    call_stack.push(current_record)

    if n <= 1:
        print(f"Base case reached: {n}")
    else:

        simulate_factorial(n - 1, f"Factorial({n})")


    call_stack.pop()


simulate_factorial(3)
```
