```python

grammar = {
    'E': ['TR'],
    'R': ['+TR', 'ε'],
    'T': ['FY'],
    'Y': ['*FY', 'ε'],
    'F': ['(E)', 'i']
}

non_terminals = list(grammar.keys())
terminals = set()

for productions in grammar.values():
    for prod in productions:
        for char in prod:
            if not char.isupper() and char != 'ε':
                terminals.add(char)

FIRST = {nt: set() for nt in non_terminals}
FOLLOW = {nt: set() for nt in non_terminals}

def compute_first(symbol):
    if symbol not in non_terminals:
        return {symbol}

    first_set = set()

    for production in grammar[symbol]:
        if production == 'ε':
            first_set.add('ε')
        else:
            for char in production:
                temp = compute_first(char)
                first_set.update(temp - {'ε'})
                if 'ε' not in temp:
                    break
            else:
                first_set.add('ε')

    return first_set

for nt in non_terminals:
    FIRST[nt] = compute_first(nt)

FOLLOW[non_terminals[0]].add('$')  # Start symbol

def compute_follow():
    changed = True
    while changed:
        changed = False

        for head, productions in grammar.items():
            for prod in productions:
                for i in range(len(prod)):
                    if prod[i] in non_terminals:
                        follow_before = FOLLOW[prod[i]].copy()

                        if i + 1 < len(prod):
                            next_symbol = prod[i+1]
                            first_next = compute_first(next_symbol)

                            FOLLOW[prod[i]].update(first_next - {'ε'})

                            if 'ε' in first_next:
                                FOLLOW[prod[i]].update(FOLLOW[head])

                        else:

                            FOLLOW[prod[i]].update(FOLLOW[head])

                        if follow_before != FOLLOW[prod[i]]:
                            changed = True

compute_follow()

print("FIRST Sets:")
for nt in FIRST:
    print(f"FIRST({nt}) = {FIRST[nt]}")

print("\nFOLLOW Sets:")
for nt in FOLLOW:
    print(f"FOLLOW({nt}) = {FOLLOW[nt]}")
```
