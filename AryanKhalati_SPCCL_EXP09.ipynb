```python
import pandas as pd

def optimize_code(instructions):

    seen_expressions = {}
    optimized_output = []

    print(f"{'Original':<20} | {'Action':<25} | {'Optimized Result'}")
    print("-" * 70)

    for target, expr in instructions:
        if expr in seen_expressions:

            replacement = seen_expressions[expr]
            action = f"Redundant! Use {replacement}"
            optimized_output.append((target, replacement))
        else:

            seen_expressions[expr] = target
            action = "New expression" # Corrected: Removed extra ')' and set a descriptive action
            optimized_output.append((target, expr)) # Added: Append the new expression to optimized_output

        print(f"{target + ' = ' + expr:<20} | {action:<25} | {optimized_output[-1][0]} = {optimized_output[-1][1]}")

    return optimized_output


code_lines = [
    ("t1", "a + b"),
    ("t2", "c * d"),
    ("t3", "a + b"),
    ("t4", "t1 + t2"),
    ("t5", "c * d")
]

print("Running CSE Optimization Simulation...\n")
final_code = optimize_code(code_lines)
```
