```python

object_code = [
    ("100", "LDA 200"),
    ("101", "ADD 201"),
    ("102", "STA 202"),
    ("103", "HLT")
]

memory = {}

for address, instruction in object_code:
    addr = int(address)
    memory[addr] = instruction

print("Memory Map After Loading:\n")
print("Address\tInstruction")

for addr in sorted(memory.keys()):
    print(f"{addr}\t{memory[addr]}")
```
