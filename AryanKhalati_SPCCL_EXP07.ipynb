```python

def precedence(op):
    if op in ('+', '-'):
        return 1
    if op in ('*', '/'):
        return 2
    return 0

def infix_to_postfix(expression):
    stack = []
    postfix = []

    for char in expression:
        if char.isalnum():
            postfix.append(char)
        elif char == '(':
            stack.append(char)
        elif char == ')':
            while stack and stack[-1] != '(':
                postfix.append(stack.pop())
            stack.pop()
        else:
            while stack and precedence(stack[-1]) >= precedence(char):
                postfix.append(stack.pop())
            stack.append(char)

    while stack:
        postfix.append(stack.pop())

    return postfix

def generate_TAC(postfix):
    stack = []
    temp_count = 1

    print("\nThree Address Code:\n")

    for char in postfix:
        if char.isalnum():
            stack.append(char)
        else:
            op2 = stack.pop()
            op1 = stack.pop()
            temp = f"t{temp_count}"
            print(f"{temp} = {op1} {char} {op2}")
            stack.append(temp)
            temp_count += 1

    return stack[-1]


expr = input("Enter expression: ").replace(" ", "")

postfix = infix_to_postfix(expr)
print("Postfix Expression:", "".join(postfix))

result = generate_TAC(postfix)

print(f"\nFinal Result Stored in: {result}")
```
