```python

assembly_code = [
    "START 100",
    "LOOP: ADD A, B",
    "SUB C, D",
    "MOVER A, B",
    "LOOP2: MULT A, C",
    "STOP",
    "END"
]

location_counter = 0
symbol_table = {}

print("Address\tLabel\tOpcode\tOperands")

for line in assembly_code:
    parts = line.replace(',', '').split()

    label = None
    opcode = None
    operands = []

    if ":" in parts[0]:
        label = parts[0].replace(":", "")
        opcode = parts[1] if len(parts) > 1 else None
        operands = parts[2:]
        symbol_table[label] = location_counter
    else:
        opcode = parts[0]
        operands = parts[1:]

    if opcode == "START":
        location_counter = int(operands[0])
        print(f"{location_counter}\t{label}\t{opcode}\t{operands}")
        continue

    print(f"{location_counter}\t{label}\t{opcode}\t{operands}")

    location_counter += 1

print("\nSymbol Table:")
print("Label\tAddress")
for sym, addr in symbol_table.items():
    print(f"{sym}\t{addr}")
```
